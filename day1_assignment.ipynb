## Task 1 — Create your data
You recorded the number of glasses of water you drank each day for one week:
`2, 5, 3, 6, 4, 7, 5`

1. Store these in a NumPy array called `water`.
2. Print the array, its `shape`, and its `size`.

```python
# YOUR CODE HERE
import numpy as np

water = np.array([2,5,3,6,4,7,5]) # creating array

print(water.shape) # shape of array
print(water.size) # size of array
```

## Task 2 — Slicing practice
From the `water` array:
1. Print the first 3 days.
2. Print the last 2 days.
3. Print only the weekend (day 6 and day 7 — positions 5 and 6).

```python
# YOUR CODE HERE
print(water[:3]) # first e days
print(water[-2:]) # last 2 days
print(water[5:]) # weekends
```

## Task 3 — Vectorized math
Each glass is 250 ml.
1. In one line, create a new array `water_ml` = glasses × 250.
2. Print the milliliters for each day.

```python
# YOUR CODE HERE
water_ml = water * 250 # liter ti milliliter

print(water_ml)
```

## Task 4 — Build a 2D array
Create a 2D array called `scores` with 2 rows and 3 columns representing two students' marks in three subjects:

Student A: 80, 75, 90
Student B: 60, 85, 70

Then print:
1. The shape of the array.
2. Student B's Maths mark (row 1, column 0).
3. The whole first column (everyone's first subject).

```python
# YOUR CODE HERE
scores = np.array([[80,75,90],
                    [60,85,70]]) # creating a multidemensional array of 2 rows and 3 columns

print(scores.shape) # shape of array
print(scores[1,0])  # score at second row and first coloumn
print(scores[:,0])  # scores in first column
```
